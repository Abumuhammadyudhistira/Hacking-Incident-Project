```python
import string

def encrypt(text,shift):
    '''
    INPUT: text as a string and an integer for the shift value.
    OUTPUT: The shifted text after being run through the Caeser cipher.
    '''
   
    # Create a normal plain alphabet
    encrypted_text = list(range(len(text))) #List of number from 0 until the number of text [0,1,2,3,..]
    
    alphabet = string.ascii_lowercase #a,b,c,d,e,f
    
    # Create a shifted version of this alphabet 
    first_half = alphabet[:shift] #['a','b','shift']
    second_half = alphabet[shift:] #['shift','d','e','f',...]
    
    shifted_alphabet = second_half+first_half #list shifted alphabet ['shift','d','e','f',...] + ['a','b','shift']
    
    
    # Use a for loop to go through each character in the original message.
    for i,letter in enumerate(text.lower()): #[(0,'t'),(1,'e'),(2,'x'),(3,'t')]
        
        if letter in alphabet: #If t,e,x,t in a,b,c,d
            
            original_index = alphabet.index(letter) # a = 0, b = 1, c= 2,...
            new_letter = shifted_alphabet[original_index] # ['shift','d','e','f',...] + ['a','b','shift'][original_index]
            encrypted_text[i] = new_letter #Huruf ke - i diisi oleh new_letter
            
        else: #If t,e,x,t not in a,b,c,d
            encrypted_text[i] = letter #Non-alphabet character tidak berubah
    
    
    return ''.join(encrypted_text)
```

```python
import string

def brute_force(message):
    """
    INPUT: A shifted message
    OUTPUT: Prints out every possible shifted message. 
            One of the printed outputs should be readable.
    """
    
    for n in range(25):
        
        # Create a normal plain alphabet
        encrypted_text = list(range(len(message))) #List of number from 0 until the number of message [0,1,2,3,..]
    
        alphabet = string.ascii_lowercase #['a','b','c','d',...]
    
        # Create a shifted version of this alphabet 
        first_half = alphabet[:n] #['a','b','n']
        second_half = alphabet[n:] #['n','d','e','f',...]
    
        shifted_alphabet = second_half+first_half #list shifted alphabet ['n','d','e','f',...] + ['a','b','n']
    
    
        # Use a for loop to go through each character in the original message.
        for i,letter in enumerate(message.lower()): #[(0,'t'),(1,'e'),(2,'x'),(3,'t')]
        
            if letter in shifted_alphabet: #If t,e,x,t in a,b,c,d
            
                original_index = shifted_alphabet.index(letter) # a = 0, b = 1, c= 2,...
                new_letter = alphabet[original_index] # ['shift','d','e','f',...] + ['a','b','shift'][original_index]
                encrypted_text[i] = new_letter #Huruf ke - i diisi oleh new_letter
            
            else: #If t,e,x,t not in a,b,c,d
                encrypted_text[i] = letter #Non-alphabet character tidak berubah
        
        
        print('Using shift value of {}'.format(n))
        print(''.join(encrypted_text))
        print('\n')
```
